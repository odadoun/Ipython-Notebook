## Fe55 emits x-rays when nucleus captures an inner electron and becomes Mn.

* The most important x-ray emissions are:

    * $K_{\alpha}$  @ 5.889 keV 
    * $K_{\beta}$  @ 6.490 keV

* $K_{\alpha}$ and $K_{\beta}$ interact in the silicon and produce a characteristic charge cloud equivalent to 1620 and 1778 electons respectively

```python
import os
import glob
 
import numpy as np
import pyfits
import pylab
 
import lsst.eotest.sensor as sensorTest
import lsst.eotest.image_utils as imutils

import lsst.afw.math as afwMath

import lsst.afw.display.ds9 as ds9
%matplotlib inline   

files = sorted(glob.glob('/var/tmp/test/*fe55_fe55*.fits'))
#amps = imutils.allAmps
amps = (1,)     # just do the first segment in each file
#
# The PsfGaussFit class processes image segments, fitting the charge
# clusters with 2D Gaussian functions.  The source detection threshold
# is give by nsig*sigma, where sigma is the stdev of the bias region.
#
fitter = sensorTest.PsfGaussFit(nsig=2)
#
# Loop over the input files, accumulating the fit data, segment by segment.
#
for item in files:
    print "processing", os.path.basename(item)
    #
    # MaskedCCD is the principal data structure for the eotest package.
    # Masks can be specified by a non-empty tuple of mask_files.
    # 
    ccd = sensorTest.MaskedCCD(item, mask_files=())
    for amp in amps:
        #print "  amp:", amp
        ds9.mtv(ccd[amp])    # display this image segment.
        fitter.process_image(ccd, amp)
    ds9.incrDefaultFrame()
fitter.write_results('psf_results.fits')
#
# Perform direct fit of DN distribution to determine the system gain
#
for amp in amps:
    results = pyfits.open('psf_results.fits')
    chiprob = np.array(results[amp].data.field('CHIPROB'), dtype=np.float)
    dn = np.array(results[amp].data.field('DN'), dtype=np.float)
    #
    # apply a chi-square probability threshold
    #
    gain = sensorTest.fe55_gain_fitter(dn, make_plot=True,
                                       title='fitted DN distribution',ylog=False)
    #print "gain for amplifier %i: %s" % (amp, gain)
    pylab.savefig('example_dn_distribution.png')
```

```python
import os
import sys
import numpy as np
import pylab
#import pyfits
import scipy.stats
import lsst.afw.detection as afwDetect
import lsst.afw.image as afwImage
import lsst.afw.math as afwMath
import lsst.eotest.image_utils as imutils
import lsst.eotest.sensor.MaskedCCD as MaskedCCD
from lsst.eotest.sensor.fe55_yield import Fe55Yield
import lsst.eotest.sensor.fe55_gain as fe55_gain
import glob

from scipy import *
from datetime import datetime
path_input="/var/tmp/test/"
fits_file_array=sorted(glob.glob(path_input+"*.fits"))
all_signals = []
all_gain = []
all_noise = []


#Loop on all the files
for ft_file in fits_file_array:
    fits_file = ft_file
    print fits_file
    ccd = MaskedCCD(fits_file)   
    #loop on all the amps
    for amp in imutils.allAmps:
    # see fe55_gain.py , Class Xrays function gain
        xrays = fe55_gain.Xrays(ccd, amp)
        signals = xrays.signals(nsig=2, max_npix=9, gain_max=10.)
        gain, noise = xrays.gain(nsig=2,max_npix= 9,gain_max=10.,make_plot=False) 
        # Make a big array with all the amps values and gain and noise for all the files
        #list of array signals (in indian file)
        all_signals.append(signals) 
        all_gain.append(gain)
        all_noise.append(noise)

#The below two lines need to be here, else xrays.gain return graphics even if make_plot=False
%matplotlib inline
import matplotlib.pyplot as plt     
f, axarr = plt.subplots(8,2,figsize=(15,30)) #by default the size for subplot is very small
 
for amp in imutils.allAmps:
    temp_signals = [] 
    temp_gain = [] 
    temp_noise = [] 

    # Make array of all the files for all the amps 
    for i in range(1,size(fits_file_array)+1):
        indice=((amp-1)+15*(i-1))
        #print amp,i,indice
        
        temp_signals.append(all_signals[indice])
        temp_gain.append(all_gain[indice])
        temp_noise.append(all_noise[indice])
       
    #Here we need to reshape the array (we have a 2D list instead of 1D!) using ravel
    signals_all_files = temp_signals #np.ravel(temp_signals)
    mean_gain = sum(np.ravel(temp_gain))/(size(fits_file_array))
    noise_std_2 = sum(np.ravel(temp_noise))
        
    flags = afwMath.MEDIAN | afwMath.STDEVCLIP

    stats = afwMath.makeStatistics(signals_all_files[0],flags)
    median = stats.getValue(afwMath.MEDIAN)
    stdev = stats.getValue(afwMath.STDEVCLIP)
    hist_nsig=10
    xmin = median - hist_nsig*stdev
    xmax = median*1785./1620. + hist_nsig*stdev
    xrange = xmin, xmax
   
    
    n,p=0,0
    if(amp < 9):
        n,p=amp-1,0
    else:
        n,p=amp-9,1
    
  #  print signals_all_files
    
    hist=axarr[n,p].hist(signals_all_files, bins=100, range=xrange, histtype='bar', color='b')
   
    #hist=pylab.hist(signals_all_files, bins=100, range=xrange, histtype='bar', color='b')
    
    x = (hist[1][1:] + hist[1][:-1])/2.
    y = hist[0] 
    ntot = sum(y)
    p0 = (ntot*0.88, median, stdev/2., ntot*0.12)
    pars, _ = scipy.optimize.curve_fit(fe55_gain.fe55_lines, x, y, p0=p0)
    kalpha_peak = pars[1]

    xx = np.linspace(x[0], x[-1], 1000)

    axarr[n,p].plot(xx, fe55_gain.fe55_lines(xx, *pars), 'r--', markersize=3, linewidth=1)
    # fe55_lines 2 Gaussian model of Mn K-alpha and K-beta lines for Fe55 tests.
    
    axarr[n,p].annotate(("K-alpha peak = %i DN\n\n" +"Gain = %.2f e-/DN\n\n" +  "Noise = %.2f e-")
              % (kalpha_peak, mean_gain, sqrt(noise_std_2)),
                (0.1, 0.7), xycoords='axes fraction')
             
```
