```python
import math
from bokeh.io import show, output_file
from bokeh.plotting import figure
from bokeh.models import GraphRenderer, StaticLayoutProvider
from bokeh.io import show, output_notebook
from bokeh.layouts import column,row
output_notebook(hide_banner=True)

N = 100
tablede=55
node_indices = list(range(N))

plot = figure(title='Multiplication Modulo', x_range=(-1.2,1.2), y_range=(-1.2,1.2))
plot.axis.visible = False
plot.xgrid.visible = False
plot.ygrid.visible = False

graph = GraphRenderer()
graph.node_renderer.data_source.add(node_indices, 'index')

graph.edge_renderer.data_source.data = dict(start=[0]*N, end=node_indices)

circ = [i*2*math.pi/N for i in node_indices]
x = [math.cos(i) for i in circ]
y = [math.sin(i) for i in circ]

pstart,pend=[],[]
for i in range(1,N+1):
    pstart.append(i)
    pend.append(i*tablede%N)
        
graph_layout = dict(zip(node_indices, zip(x, y)))
graph.layout_provider = StaticLayoutProvider(graph_layout=graph_layout)

graph.edge_renderer.data_source.data['start']=pstart
graph.edge_renderer.data_source.data['end']=pend
plot.renderers.append(graph)
show(plot)
```
